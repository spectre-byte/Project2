### 5 Data Processing

#### 5.1 Scaling the Data Frame

```python
scaled_df = scalar.fit_transform(df)
```

#### 5.2 Dimensionality Reduction
- Converting the DataFrame into 2D DataFrame for better visualization.

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

### 6 HyperParameter Tuning
- Finding 'k' value by Elbow Method

```python
inertia = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

### 7 Model Building using K-means Clustering

```python

kmeans_model=KMeans(3)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

### 8 Visualizing the Clustered Dataframe

```python
plt.figure(figsize=(9,9))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','blue'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

#### 8.1 Find all clusters

```python
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[df.columns])
```

#### 8.2 Inverse transform the data

```python
cluster_centers = scalar.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

#### 8.3 Creating a target column "Cluster" for storing the cluster segment

```python
cluster_df = pd.concat([df,pd.DataFrame({'Cluster':kmeans_model.labels_})],axis=1)
cluster_df
```

### 9  Saving the K-means Clustering model and the data with cluster label

```python
#Saving Scikitlearn models
import joblib
joblib.dump(kmeans_model, "kmeans_model.pkl")
```

### 10 Training and Testing the model accuracy using decision tree

```python
#Split Dataset
X = cluster_df.drop(['Cluster'],axis=1)
y= cluster_df[['Cluster']]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
X_test
```

### 11 Training the model using Decision Tree for the future predictions

```python
#Decision_Tree
model= DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#Confusion_Matrix
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Save the Decision Tree Model

```python
import pickle
filename = 'final_model.sav'
pickle.dump(model, open(filename, 'wb'))
# some time later...
 
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)
print(result,'% Acuracy')
```
